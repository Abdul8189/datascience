```python
import pandas as pd
```

```python
dataset=pd.read_csv('Iris.csv')
```

```python
dataset.head(3)
```

```python
dataset.drop(columns=['Id','Species'], inplace=True)
```

```python
dataset.head(3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
import scipy.cluster.hierarchy as sc
```

```python
sc.dendrogram(sc.linkage(dataset,method='single',metric='euclidean'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac=AgglomerativeClustering(n_clusters=2,linkage='single')
```

```python
dataset['predict']= ac.fit_predict(dataset)
```

```python
sns.pairplot(data=dataset,hue='predict')
plt.show()
```
