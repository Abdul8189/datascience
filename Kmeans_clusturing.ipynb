```python
import pandas as pd
```

```python
dataset=pd.read_csv('Iris.csv')
```

```python
dataset.head(3)
```

```python
dataset.drop(columns=['Id','Species'],inplace=True)
```

```python
dataset.head(3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.scatterplot(data=dataset)
plt.show()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (2,21):
  km=KMeans(n_clusters=i,init='k-means++')
  km.fit(dataset)
  wcss.append(km.inertia_)
```

```python
from matplotlib import markers
plt.plot([i for i in range (2,21)],wcss,marker='o')
plt.xticks([i for i in range (2,21)])
plt.grid(axis='x')
plt.show()
```

```python
kns=KMeans(n_clusters=3)
```

```python
kns.fit(dataset)
kns.fit_predict(dataset)
```

```python
dataset['species']=kns.fit_predict(dataset)
```

```python
dataset.head(3)
```

```python
sns.pairplot(data=dataset,hue='species')
plt.show()
```
