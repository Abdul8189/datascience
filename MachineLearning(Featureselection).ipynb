```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
dataset=pd.read_csv('diabetes.csv')
dataset.head(10)
```

```python
dataset.shape
```

```python
dataset.info()
```

```python

dataset['BMI']=dataset['BMI'].astype('int64')
dataset['DiabetesPedigreeFunction']=dataset['DiabetesPedigreeFunction'].astype('int64')
```

```python
x=dataset.iloc[:,:-1]
y=dataset['Outcome']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
fs=SequentialFeatureSelector(lr,k_features=6,forward=True)
fs.fit(x,y)
```

```python
fs.feature_names
```

```python
fs.k_feature_names_
```

```python
fs.k_score_
```

```python
8-0.7708683473389355
```

```python

fs=SequentialFeatureSelector(lr,k_features=8,forward=False)
fs.fit(x,y)
```

```python
fs.k_feature_names_
```

```python
fs.k_score_
```
