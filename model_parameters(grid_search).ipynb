```python
import pandas as pd
```

```python
dataset= pd.read_csv('cgpa.csv')
```

```python
dataset.head(3)
```

```python
x=dataset.iloc[:,:-1]
y=dataset['package']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import  DecisionTreeRegressor
```

```python
dc= DecisionTreeRegressor(criterion='friedman_mse',splitter='random',max_depth=4)
```

```python
dc.fit(x_train,y_train)
```

```python
dc.score(x_test,y_test)
```

```python
dc.score(x_train,y_train)
```

```python
from sklearn.model_selection import  GridSearchCV,RandomizedSearchCV
```

```python
df = {"criterion":["squared_error", "friedman_mse", "absolute_error", "poisson"],
      "splitter":["best", "random"],
      "max_depth":[i for i in range(2,20)]}
```

```python
gr=GridSearchCV(DecisionTreeRegressor(),param_grid=df)
gr.fit(x_train,y_train)
```

```python
gr.best_params_
```

```python
gr.best_score_
```

```python
rd=RandomizedSearchCV(DecisionTreeRegressor(),param_distributions=df,n_iter=20)
rd.fit(x_train,y_train)
```

```python
rd.best_params_
```

```python
rd.best_score_
```
