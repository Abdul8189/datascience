```python
import pandas as pd
```

```python
dataset=pd.read_csv('placement.csv')
```

```python
dataset.head(3)
```

```python
new_data=dataset.head(10)
```

```python
x=new_data.iloc[:,:-1]
y=new_data['placed']
```

```python
print(y.shape)
```

```python
new_data
```

```python
from sklearn.model_selection import LeaveOneOut,LeavePOut,KFold,StratifiedKFold
```

```python
lo=LeaveOneOut()
for train,test in lo.split(x,y):
  print(train,test)
```

```python
lo1=LeavePOut(p=2)
for train,test in lo1.split(x,y):
  print(x,y)
```

```python

lo2=KFold(n_splits=5)

for train,test in lo2.split(x,y):
  print(train,test)
```

```python
lo3=StratifiedKFold(n_splits=5)
for train,test in lo3.split(x,y):
  print(train,test)
```
