```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset=pd.read_csv("placement.csv")
```

```python
dataset.head(3)
```

```python
dataset.rename(columns={'resume_score':'score'},inplace=True)
```

```python
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(x='cgpa',y='score',data=dataset,hue='placed')
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset['placed']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
sns.kdeplot(data=dataset['cgpa'])
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
```

```python
gb=GaussianNB()
```

```python
gb.fit(x_train,y_train)
```

```python
gb.score(x_test,y_test)
```

```python
mb=MultinomialNB()
```

```python
mb.fit(x_train,y_train)
```

```python
mb.score(x_test,y_test)
```

```python
br=BernoulliNB()
```

```python
br.fit(x_train,y_train)
```

```python
br.score(x_test,y_test)
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gb)
plt.show()
```

```python
plt.figure(figsize=(5,4))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mb)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=br)
plt.show()
```
