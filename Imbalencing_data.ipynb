```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('ads.csv')
```

```python
dataset.head(10)
```

```python
dataset['Purchased'].value_counts()
```

```python
x= dataset.iloc[:,:-1]
```

```python
y= dataset['Purchased']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
lr.predict([[46,	26000	]])
```

```python
from imblearn.under_sampling import  RandomUnderSampler
```

```python
rm=RandomUnderSampler()
```

```python
rm.fit(x,y)
```

```python
rm.fit_resample(x,y)
```

```python
ru_x,ru_y=rm.fit_resample(x,y)
```

```python
ru_x
```

```python
ru_y
```

```python
ru_x.value_counts()
```

```python
ru_y.value_counts()
```

```python
x1_train,x1_test,y1_train,y1_test=train_test_split(ru_x,ru_y,test_size=0.2)
```

```python
lr1=LogisticRegression()
```

```python
lr1.fit(x1_train,y1_train)
```

```python
lr1.predict([[45,35000]])
```

```python
lr1.score(x1_test,y1_test)
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ro=RandomOverSampler()
```

```python
ro.fit_resample(x,y)
```

```python
xd1,yd1=ro.fit_resample(x,y)
```

```python
xd1.value_counts()
```

```python
yd1.value_counts()
```

```python
x2_train,x2_test,y2_train,y2_test=train_test_split(xd1,yd1,test_size=0.2)
```

```python
lr2=LogisticRegression()
```

```python
lr2.fit(x2_train,y2_train)
```

```python
lr2.score(x2_test,y2_test)
```

```python
lr2.predict([[45,26000]])
```
